```python
# Importing Dependencies
import pandas as pd
import os
from sqlalchemy import create_engine
```

# Extract - Storing CSV into DataFrame

```python
# Creating csv path and loading first data in Pandas DataFrame
csv_path = ("World-population-by-countries-dataset.csv")
wrld_pop_df = pd.read_csv(csv_path)
wrld_pop_df.head()
```

```python
# Creating csv path and loading first data in Pandas DataFrame
csv_path = ("unemployment-analysis.csv")
unemp_rate_df = pd.read_csv(csv_path)
unemp_rate_df.head()
```

# Data Cleaning and Transformation

```python
# Renaming Country Name column
mapping = {wrld_pop_df.columns[0]:'country_name'}
transformed_data = wrld_pop_df.rename(columns=mapping)
transformed_data.head()
```

```python
# Renaming Country Name column
mapping = {unemp_rate_df.columns[0]:'country_name'}
renamed_data = unemp_rate_df.rename(columns=mapping)
# renamed_data
```

```python
# Creating a filtered dataframe from specific columns

transformed_data_cols = ['country_name', '2011', '2012', '2013', '2014', '2015', 
                 '2016', '2017', '2018', '2019', '2020', '2021']

world_pop_df = transformed_data[transformed_data_cols].copy()

# Clean the data by dropping duplicates and setting the index
world_pop_df.drop_duplicates()

# Setting the Country Name column as index
# world_pop_df.set_index("country_name", inplace=True)

world_pop_df.head()
```

```python
# Creating a filtered dataframe from specific columns

renamed_data_cols = ['country_name', '2011', '2012', '2013', '2014', '2015', 
                 '2016', '2017', '2018', '2019', '2020', '2021']

unemp_rates_df = renamed_data[renamed_data_cols].copy()


# Clean the data by dropping duplicates and setting the index
unemp_rates_df.drop_duplicates()

# Setting the Country Name column as index
#unemp_rate_transformed.set_index("Country Name", inplace=True)

unemp_rates_df.head()
```

```python
# Generate a new dataframe re-orienting the original dataframe using pd.melt
df = pd.melt(unemp_rates_df, id_vars=['country_name'], value_vars=['2011', '2012', '2013','2014', '2015','2016','2017','2018','2019','2020'])
df.head()
```

# Connecting to local database

```python
# Connect to SQLLite database
engine = create_engine('sqlite:///data/unemployment.db', echo=False)
df.to_sql('census',con=engine,if_exists='replace')
```
