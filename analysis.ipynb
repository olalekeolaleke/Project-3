```python
# Renaming Country Name column
mapping = {wrld_pop_df.columns[0]:'Country'}
transformed_data = wrld_pop_df.rename(columns=mapping)
#transformed_data.head()
```

```python
# Renaming Country Name column
mapping = {unemp_rate_df.columns[0]:'Country'}
renamed_data = unemp_rate_df.rename(columns=mapping)
#renamed_data
```

```python
# Creating a filtered dataframe from specific columns

transformed_data_cols = ['Country', '2015', 
                 '2016', '2017', '2018', '2019', '2020', '2021']

world_pop_df = transformed_data[transformed_data_cols].copy()

# Clean the data by dropping duplicates 
world_pop_df.drop_duplicates()

world_pop_df.head()
```

```python
# Creating a filtered dataframe from specific columns

renamed_data_cols = ['Country', '2015', 
                 '2016', '2017', '2018', '2019', '2020', '2021']

country_unemployment = renamed_data[renamed_data_cols].copy()


# Clean the data by dropping duplicates and setting the index
country_unemployment.drop_duplicates()

# Setting the Country Name column as index
#unemp_rate_transformed.set_index("Country Name", inplace=True)

country_unemployment.head()
```
